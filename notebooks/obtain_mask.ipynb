This notebook demonstrates the mask obtaining procedure from `photopea` edited pictures. 

```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
im = Image.open("pinhole_00000_edited.tga")
print(im.format, im.size, im.mode)
```

```python
plt.imshow(im)
```

```python
import numpy as np
im_data = np.asarray(im)
mask_data = im_data[:,:,3]

plt.imshow(mask_data)
```

```python
mask = Image.fromarray(mask_data)
mask.save('mask_pinhole0.jpg')
```

```python
def tga_to_mask(im_path: str) -> np.ndarray:
    im = Image.open(im_path)
    im_data = np.asarray(im)
    mask_data = im_data[:,:,3]
    mask = Image.fromarray(mask_data)
    return mask
```

```python
im_path = "pinhole_duplicate0_00000_edited.tga"
output_path = "mask_pinhole_duplicate0.jpg"
mask = tga_to_mask(im_path)
```

```python
plt.imshow(mask)
```

```python
mask.save(output_path)
```
